```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Define data

```python
students = np.array(["Ilia", "Vlad", "Katya", "Feodosiy", "Eva"])
attendance = np.array([100, 98, 100, 99, 100])
marks = np.array([100, 97, 99, 100, 100])
```

```python
data_dict = {
    "Name": students,
    "Attendance": attendance,
    "Mark": marks
}
```

```python
df = pd.DataFrame(data_dict)
```

```python
plt.scatter(df['Attendance'], df['Mark'], color="red")
plt.title("Marks and Attendance")
plt.xlabel("% of classes attended")
plt.ylabel("Mark")
plt.show()
```

# Conclusion
- It seems there is a positive correlation between attendance and marks :)
