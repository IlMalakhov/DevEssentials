```python
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
crashes = sns.load_dataset('car_crashes')
```

# Drawing the first plot

```python
sns.relplot(
    data=crashes,
    x="total", y="speeding", size="alcohol", hue="not_distracted"
            )
```

## Insights:
- Positive Correlation
    - Between Total Crashes and Speeding: 
        As the total number of crashes increases, the number of speeding-related crashes also tends to increase. This is evident from the upward trend in the points.
    - Alcohol Involvement:
        Larger points (indicating higher alcohol involvement) tend to cluster toward the middle and upper right of the graph, suggesting that as the total number of crashes and speeding-related crashes increase, alcohol involvement also tends to be higher.
- Distraction Levels:
    - Darker points (indicating more non-distracted crashes) are more frequent in the upper right of the graph, suggesting that in cases with higher speeding and total crashes, distraction is less of a factor.
    - Conversely, lighter points (fewer non-distracted crashes) are more common in the lower left, indicating that distracted behavior might be more common in smaller crashes.


# Drawing the second plot

```python
sns.relplot(data = crashes,
            x='total', 
            y = 'ins_premium', 
            hue = 'no_previous', 
            size = 'ins_losses'
            )
```

## Insights:
- Clustering: There is a noticeable concentration of data points around the lower total crash numbers and premiums, suggesting a common occurrence of lower fatal crash numbers and lower premiums.


# Relationship between alcohol and speeding

```python
sns.lmplot(
    data=crashes,
    x='speeding',
    y='alcohol'
            )
```

## Insights:
- Co-occurrence: States with higher speeding percentages tend to also have a higher percentage of alcohol-impaired drivers involved in fatal accidents. This co-occurrence may suggest that these two behaviors often contribute together to road fatalities.
- Variability: Theres also notable variability in some areas where states with similar speeding rates exhibit quite different alcohol impairment rates, which could point to the influence of other factors (e.g., law enforcement, culture, or specific driving behaviors).
