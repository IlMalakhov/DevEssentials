# Importing matplotlib and statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    # ++
import statistics as st            # ++
```

```python
incomes = [0, 0, 3, 120, 403, 1009, 20199, 19200, 50125, 90263]
expenses = [20000, 20000, 40000, 40000, 50000, 90000, 100000, 120000, 120000, 150000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
```

```python
np_incomes = np.array(incomes)
np_incomes_taxed = np_incomes * 0.7
np_expenses = np.array(expenses)
np_savings = np_incomes_taxed - np_expenses    # Added savings
np_incomes, np_incomes_taxed, np_expenses
```

```python
df_data_dict = {
    'Taxed Incomes' : np_incomes_taxed,
    'Expenses' : np_expenses,
    'Savings' : np_savings    # Added savings
}
df_income_expenses = pd.DataFrame(df_data_dict, months)
df_income_expenses
```

## First quarter

```python
first_quarter = df_income_expenses.loc['Jan':'Mar']
first_quarter
```

## Second quarter

```python
second_quarter = df_income_expenses.loc['Apr':'Jun']
second_quarter
```

# Using matplotlib

## Plot

```python
plt.plot(months, np_incomes_taxed, color='black', marker='.', markersize = 8)
plt.title('Income by month')
plt.show()
```

## Bar chart

```python
plt.bar(months, np_savings, color='red', width = 0.7)
plt.title('Savings by month 0_0')
plt.show()
```

## Pie Chart
Pie Charts in matplotlib don't natively support negative values so we have to make `np_savings` absolute. I found `np.abs`, but could also `losses = np_savings * -1`

```python
losses = np.abs(np_savings)
plt.pie(losses, labels=months)
plt.title('Losses for each month')
plt.show()
```

# Average income for quarters

```python
quarters = ['1st', '2nd', '3rd', '4th']
quarter_averages = [st.mean(np_incomes_taxed[0:3]),
                    st.mean(np_incomes_taxed[3:6]),
                    st.mean(np_incomes_taxed[6:9]),
                    np_incomes_taxed[9]]            # There are only ten months
```

```python
plt.plot(quarters, quarter_averages, color='black', marker = 'o')
plt.title('Average income for quarters')
plt.show()
```
